```
train <- read.csv("train.csv")
```

```
train
```

```
train %>% count(Embarked)
```

```
count(Embarked)
```

```
train %>%
count(Embarked) %>%
ggplot(aes(x = Embarked ,y=n ,fill=Embarked))+
geom_col(, fill = c("#003f5c", "#444e86", "#955196", "#dd5182")) +
geom_text(aes(label=n), vjust=-0.25)+
labs(title="不同港口之人數比",x="港口",y="數量")+ theme(legend.position="none")
```

```
train %>%
count(Sex) %>%
ggplot(aes(x=Sex,y=n))+
geom_col(, fill = c("#ff6e54", "#ffa600"))+
labs(title="不同性別之人數比",x="性別",y="數量")+ theme(legend.position="none")
```

```
train  %>%
count(Age) %>%
ggplot(aes(x=Age,y=n))+
geom_col(, fill =c("#dd5182"))+
labs(title="不同年齡之人數比",x="年齡",y="數量")+ theme(legend.position="none")
```

```
install.packages("ggridges")
```

```
library(ggridges)
```

```
train  %>%
ggplot(aes(x=Age,y=Embarked,fill=Embarked)) +
geom_density_ridges()+
labs(title="不同年齡在不同港口上岸之人數比",x="年齡",y="港口")+ theme(legend.position="none")
```

```
train %>%
mutate(Embarked = fct_reorder(Embarked,Age,median, na.rm= TRUE))%>%
ggplot(aes(Embarked,Age))+
geom_point()+
geom_boxplot()+
coord_flip()+
labs(title="不同年齡在不同港口上岸之人數比",x="年齡",y="港口")+ theme(legend.position="none")
```

```
install.packages("ggbeeswarm")
library(ggbeeswarm)
train %>%
ggplot(aes(
x = Pclass,
y = Age,
colour = Sex,
alpha = Sex
)) +
ggbeeswarm::geom_beeswarm() +
coord_flip() +
scale_color_manual(values = c("#BB1288", "#5867A6")) +
scale_alpha_manual(values = c(1, .4)) +
theme_minimal() +
theme(legend.position = "top") +
labs(
title = "各船票等級性別分析"
,
subtitle = "鐵達尼號乘客數據"
,

colour = "Sex",
alpha = "Sex",
x = "船票等級"
,
y = "年齡"
)
```
