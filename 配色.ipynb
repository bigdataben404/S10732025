```
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1718,color=Position)  #X=FieldGoalsAttempted(FAG,投籃出手次數)   Y=TotalPoints(總得分)       #color=Position 用守備位置著色
```

```
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1718, 
geom = c("point", "smooth")) #加點與漸進線
```

```
qplot(TotalPoints, data = NBA1718, fill = Position) #只有單變量TotalPoint的直方圖 
```

```
qplot(FieldGoalsAttempted, TotalPoints, data = NBA1718, facets= . ~ Position)#用守備位置Position分群畫圖(直向)
```

```
qplot( TotalPoints , data = NBA1718, facets=  . ~ Position, binwidth=100 )#
```
